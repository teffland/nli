### What is the distribution of hypothesis and premise lengths?

Almost all premises < 50 tokens and premises are even shorter; < 30

```python
train_hlens = np.array([len(h) for h in train_data['hs']])
train_plens = np.array([len(p) for p in train_data['ps']])
dev_hlens = np.array([len(h) for h in dev_data['hs']])
dev_plens = np.array([len(p) for p in dev_data['ps']])
fig, axs = plt.subplots(2,2, figsize=(12,9))
bins = np.linspace(0,100,11)

axs[0,0].hist(train_hlens, bins=bins)
axs[0,0].set_xticks(bins)
axs[0,0].set_title('Train Hypothesis Token Counts')

axs[0,1].hist(train_plens, bins=bins)
axs[0,1].set_xticks(bins)
axs[0,1].set_title('Train Premise Token Counts')

axs[1,0].hist(dev_hlens, bins=bins)
axs[1,0].set_xticks(bins)
axs[1,0].set_title('Dev Hypothesis Token Counts')

axs[1,1].hist(dev_plens, bins=bins)
axs[1,1].set_xticks(bins)
axs[1,1].set_title('Dev Premise Token Counts')
```
